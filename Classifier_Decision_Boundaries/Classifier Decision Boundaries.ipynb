This project loads 3 different datasets and generates classifiers for the elements within each dataset using the Logistic Regression, Guassian Naive Bayes, and K-Nearest Neighbors algorithms. **2)Use logistic regression to fit() and score() a binary classifier for dataset 1. How accurate are the model’s predictions?**

**Dataset 1 Decision Boundaries**

```python
#Setup for dataset 1
data1_min1, data1_max1 = data1[0].min()-1, data1[0].max()+1
data1_min2, data1_max2 = data1[1].min()-1, data1[1].max()+1
data1_x1grid = np.arange(data1_min1, data1_max1, 0.1)
data1_x2grid = np.arange(data1_min2, data1_max2, 0.1)
data1_xx, data1_yy = np.meshgrid(data1_x1grid, data1_x2grid)
data1_r1, data1_r2 = data1_xx.flatten(), data1_yy.flatten()
data1_r1, data1_r2 = data1_r1.reshape((len(data1_r1), 1)), data1_r2.reshape((len(data1_r2), 1))
data1_grid = np.hstack((data1_r1,data1_r2))
#End setup for dataset 1

#Logistic Regression Model Boundaries Dataset 1
logreg_data1_yhat = logreg_data1.predict(data1_grid)
logreg_data1_zz = logreg_data1_yhat.reshape(data1_xx.shape)
plt.contourf(data1_xx, data1_yy, logreg_data1_zz, cmap='Paired')
plt.title('Logistic Regression Model Boundaries Dataset 1')
plt.xlabel('X-1')
plt.ylabel('X-2')
plt.scatter(data1[0],data1[1], c=data1[2])
```

```python
#Gaussian Naive Bayes Model Boundaries Dataset 1
gnb_data1_yhat = gnb_data1.predict(data1_grid)
gnb_data1_zz = gnb_data1_yhat.reshape(data1_xx.shape)
plt.contourf(data1_xx, data1_yy, gnb_data1_zz, cmap='Paired')
plt.title('Gaussian Naive Bayes Model Boundaries Dataset 1')
plt.xlabel('X-1')
plt.ylabel('X-2')
plt.scatter(data1[0],data1[1], c=data1[2])
```

```python
#K-Nearest Neighbors Model Boundaries Dataset 1
knn_data1_yhat = knn_data1.predict(data1_grid)
knn_data1_zz = knn_data1_yhat.reshape(data1_xx.shape)
plt.contourf(data1_xx, data1_yy, knn_data1_zz, cmap='Paired')
plt.title('K-Nearest Neighbors Model Boundaries Dataset 1')
plt.xlabel('X-1')
plt.ylabel('X-2')
plt.scatter(data1[0],data1[1], c=data1[2])
```

Each of the classifier algorithms produce boundaries that correctly predict all of the elements within our dataset. The Logistic Regression and Gaussian Naive Bayes algorithms both produce what appears to be a straight line while the K-Nearest Neighbors algorithm produces a line that is similar with some slight curves.
