This project pulls student meeting attendance data from a csv file, generates the values for the quartile ranges, and outputs outliers per week to the console in text format. This project relies on solely the CSV module and uses the Python standard library for all calculations. The functions defined within this project were designed for the given data set. To accomdate for other data sets, changes would need to be made to the quartile functions for different sized lists of data.


In order for project to run, CSV file containing data must be located in same directory.


```python
import csv

# Lists to store CSV Information
student_names = []
week_one_times = []
week_two_times = []
week_three_times = []
week_one_times_sorted = []
week_two_times_sorted = []
week_three_times_sorted = []


# load csv and store information in lists
def load_csv(file):
    with open(file, newline='') as csvfile:
        student_data = csv.reader(csvfile, delimiter=',', quotechar='|')
        next(student_data)  # skip first line
        for row in student_data:
            student_names.append(row[0])
            week_one_times.append(int(row[1]))
            week_two_times.append(int(row[2]))
            week_three_times.append(int(row[3]))


# stores unsorted data as new sorted list
def sort_data(unsorted_list, sorted_list):
    for x in range(0, len(unsorted_list)):
        sorted_list.append(int(sorted(unsorted_list)[x]))

        
def quartile_one(sorted_list):
    return (sorted_list[int((len(sorted_list))/4-1)] + sorted_list[int((len(sorted_list))/4)]) / 2


def quartile_two(sorted_list):
    return (sorted_list[int((2*len(sorted_list))/4-1)] + sorted_list[int((2*len(sorted_list))/4)]) / 2


def quartile_three(sorted_list):
    return (sorted_list[int((3*len(sorted_list))/4-1)] + sorted_list[int((3*len(sorted_list))/4)]) / 2


# Calculates lower turkey fence for outliers using k = 1.5
def lower_turkey_fence(q1, q3):
    return q1 - ((q3 - q1)*1.5)


# Calculates upper turkey fence for outliers using k = 1.5
def upper_turkey_fence(q1, q3):
    return q3 + ((q3 - q1)*1.5)


# Consolidation of calculations for week one - week three
def tardy():
    sort_data(week_one_times, week_one_times_sorted)
    sort_data(week_two_times, week_two_times_sorted)
    sort_data(week_three_times, week_three_times_sorted)

    # =============================================================
    # Week One
    # =============================================================
    week_one_quartile_one = quartile_one(week_one_times_sorted)
    week_one_quartile_two = quartile_two(week_one_times_sorted)
    week_one_quartile_three = quartile_three(week_one_times_sorted)

    week_one_lower = lower_turkey_fence(week_one_quartile_one, week_one_quartile_three)
    week_one_upper = upper_turkey_fence(week_one_quartile_one, week_one_quartile_three)

    print("Week One Outliers:")
    print("Name : Attendance Time")
    for x in range(0, len(week_one_times)):
        if week_one_times[x] < week_one_lower or week_one_times[x] > week_one_upper:
            print(student_names[x] + " : " + str(week_one_times[x]))
    print()
    
    # =============================================================
    # Week Two
    # =============================================================
    week_two_quartile_one = quartile_one(week_two_times_sorted)
    week_two_quartile_two = quartile_two(week_two_times_sorted)
    week_two_quartile_three = quartile_three(week_two_times_sorted)

    week_two_lower = lower_turkey_fence(week_two_quartile_one, week_two_quartile_three)
    week_two_upper = upper_turkey_fence(week_two_quartile_one, week_two_quartile_three)

    print("Week Two Outliers: ")
    print("Name : Attendance Time")
    for x in range(0, len(week_two_times)):
        if week_two_times[x] < week_two_lower or week_two_times[x] > week_two_upper:
            print(student_names[x] + " : " + str(week_two_times[x]))
    print()

    # =============================================================
    # Week Three
    # =============================================================
    week_three_quartile_one = quartile_one(week_three_times_sorted)
    week_three_quartile_two = quartile_two(week_three_times_sorted)
    week_three_quartile_three = quartile_three(week_three_times_sorted)

    week_three_lower = lower_turkey_fence(week_three_quartile_one, week_three_quartile_three)
    week_three_upper = upper_turkey_fence(week_three_quartile_one, week_three_quartile_three)

    print("Week Three Outliers:")
    print("Name : Attendance Time")
    for x in range(0, len(week_three_times)):
        if week_three_times[x] < week_three_lower or week_three_times[x] > week_three_upper:
            print(student_names[x] + " : " + str(week_three_times[x]))


if __name__ == "__main__":
    load_csv('participants.csv')
    tardy()

```
